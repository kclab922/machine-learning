```python
dataset = pd.read_csv('KMeansData.csv')
dataset.head()
```

```python
# 비지도학습은 종속변수를 주지 않고 학습시키는 것이므로, y변수는 불필요.
X = dataset.iloc[:, :].values
# X = dataset.values
# X = dataset.to_numpy()
X[:5]
```

### 데이터 시각화(x축-y축 범위 통일)

```python
# 점들 간의 거리 산정 시 x축과 y축의 범위가 서로 다르면 혼동이 발생할 수 있음 => 통일 필요
plt.scatter(X[:,0], X[:, 1]) # X축 hours, y축 score
plt.title('Score by hours')
plt.xlabel('hours')
plt.xlim(0, 100)
plt.ylabel('score')
plt.ylim(0, 100)
plt.show()
```

### Feature Scaling

```python
# StandardScaler를 import하기 전에 불러와야 함
# OMP_NUM_THREADS
# OpenMP 라이브러리가 병렬 처리를 위해 사용하는 스레드 수를 설정하는 데 사용되는 환경 변수.
# 기본적으로 1로 설정하면 여러 스레드를 사용하여 시스템 과부하가 발생할 수 있는 일부 경우에 도움이 될 수 있지만 항상 최적의 성능 솔루션이 아닐 수도 있음.
# 일반적으로 최상의 성능을 얻으려면 애플리케이션의 특정 요구 사항에 따라 이 변수를 조정하는 것이 좋음.
import os
os.environ['OMP_NUM_THREADS'] = '1'
```

```python
# x축과 y축의 범주를 통일해줌
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X[:5]
```

### 데이터 시각화 (스케일링된 데이터)

```python
plt.figure(figsize=(3,3))
plt.scatter(X[:,0], X[:, 1]) # X축 hours, y축 score
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

### Finding optimal K (Elbow Method)

```python
from sklearn.cluster import KMeans

# inertia: 한 클러스터 내의 데이터들이 얼마나 가깝게 모여있는지를 보여주는 지표
inertia_list = []

for i in range(1, 11): # K값(클러스터의 개수)을 1~10까지로 설정
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, random_state=0) 
    # init 설정: 중심점 선정을 무작위 랜덤이 아닌 먼거리 점들부터 하기 위함
    # n_init 설정: 초기 중심점을 선택하는 시도 횟수 (현 버전에서는 10이 기본값이나, 이후 버전에서는 auto가 기본값이 될 예정이므로, 기본값 쓰려면 10을 설정하라는 경고뜸)
    kmeans.fit(X)
    inertia_list.append(kmeans.inertia_) 
    # K값(클러스터 개수)가 바뀔 때마다 각 지점으로부터 클러스터의 중심(centroid)까지의 거리의 제곱의 합을 리스트에 누적
    
plt.plot(range(1,11), inertia_list)
plt.title('Elbow Method')
plt.xlabel('n_clusters')
plt.ylabel('inertia')
plt.show()
```

### optimal K(4)로 K-Means 학습

```python
# 기울기가 급격히 완만해지는 K값(클러스터의 개수)가 optimal K. 여기서는 4
K = 4
```

```python
kmeans = KMeans(n_clusters=4, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X) # 학습 후 X반영 시 예측되는 값, 각 데이터가 어느 클러스터에 속하는지
y_kmeans # 클러스터가 총 4개이므로, 각 데이터는 0123 에 속함.
```

### 데이터 시각화 (optimal K)

```python
# 각 클러스터의 중심점 (centroid) 좌표 (스케일링된 데이터 기준)
# 현재 클러스터의 개수가 4개이므로, 4개의 좌표가 도출됨
centers = kmeans.cluster_centers_
centers 
```

```python
for cluster in range(K):
    plt.scatter(X[y_kmeans == cluster, 0], X[y_kmeans == cluster, 1], s=100, edgecolor='black') 
    # 각 데이터 출력. 0: hour, 1: score / y_kmeans == cluster: 인덱스 번호
    
    plt.scatter(centers[cluster, 0], centers[cluster, 1], s=300, edgecolor='black', color='y', marker='s')
    # 중심점 노란 네모
    
    plt.text(centers[cluster, 0], centers[cluster, 1], cluster, ha='center', va='center')
    # 클러스터 번호 출력
    
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

### 데이터 시각화 (스케일링 원상복구)

```python
# Feature Scaling된 데이터를 다시 원상복구
# 스케일링된 X를 원상복구
X_org = sc.inverse_transform(X)
X_org[:5]
```

```python
# 중심점 원상복구
centers_org = sc.inverse_transform(centers)
centers_org
```

```python
for cluster in range(K):
    plt.scatter(X_org[y_kmeans == cluster, 0], X_org[y_kmeans == cluster, 1], s=100, edgecolor='black') 
    # 각 데이터 출력. 0: hour, 1: score / y_kmeans == cluster: 인덱스 번호
    
    plt.scatter(centers_org[cluster, 0], centers_org[cluster, 1], s=300, edgecolor='black', color='y', marker='s')
    # 중심점 노란 네모
    
    plt.text(centers_org[cluster, 0], centers_org[cluster, 1], cluster, ha='center', va='center')
    # 클러스터 번호 출력
    
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```
